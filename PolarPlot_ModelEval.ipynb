```python
import pandas as pd
import numpy as np
import plotly.express as px

```

```python
data=pd.read_excel('/content/Model_Results.xlsx')
```

```python
data=data[['Model ', 'Mean_AbsErr', 'Mean_squre_error', 'r2_score', 'Mean_comb',
       'mean_squared_error_comb', 'r2_score_comb', 'Dice-similarity_coeff']]
```

```python
data=data.T
```

```python
data=data.dropna(axis=1)
```

```python
data.rename(columns=data.iloc[0],inplace=True)
```

```python
data.reset_index(inplace=True)
data=data.loc[1:]

```

```python
fig = px.line_polar(data, r=data['dV3_resent50_mc3'].tolist(), theta=data['index'].tolist(), line_close=True,title="dV3_resent50_mc3")

fig.update_traces(fill='toself')
fig.show()
```

```python
import plotly.graph_objects as go
import plotly.offline as pyo


categories = data['index'].tolist()
categories = [*categories, categories[0]]

model_1 = data['dV3_resent50_mc3'].tolist()
model_2 = data['dV3_resent50_r2dplus1'].tolist()
model_3 = data['dV3_resent50_r3d_18'].tolist()
model_1 = [*model_1, model_1[0]]
model_2 = [*model_2, model_2[0]]
model_3 = [*model_3, model_3[0]]


fig = go.Figure(
    data=[
        go.Scatterpolar(r=model_1, theta=categories, name='dV3_resent50_mc3'),
        go.Scatterpolar(r=model_2, theta=categories, name='dV3_resent50_r2dplus1'),
        go.Scatterpolar(r=model_3, theta=categories, name='dV3_resent50_r3d_18')
    ],
    layout=go.Layout(
        title=go.layout.Title(text='Model comparison'),
        polar={'radialaxis': {'visible': True}},
        showlegend=True
    )
)

pyo.plot(fig)
```
